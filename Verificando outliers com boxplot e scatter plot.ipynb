```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
base = pd.read_csv('credit_data.csv')
```

```python
base.head()
```

```python
# Nesse caso, removeremos todos os dados nan do dataset
base = base.dropna()
```

```python
# Outliers da idade
plt.boxplot(base.iloc[:,2], showfliers=True)
```

```python
# É possível verificar, pelo boxplot, que temos idades menores do que zero
outliers_age = base[base['age'] < 0]
outliers_age
```

```python
# Outliers do empréstimo
plt.boxplot(base.iloc[:,3])
```

```python
# Observando os outliers do empréstimo
outliers_loan = base[base.loan > 13400]
outliers_loan
```

```python
# Verificando outliers utilizando income x age
plt.scatter(base.iloc[:,1], base.iloc[:,2])
plt.ylabel('Age')
plt.xlabel('Income')
```

```python
# Verificando outliers utilizando income x loan
plt.scatter(base.iloc[:,1], base.iloc[:,3])
plt.ylabel('Loan')
plt.xlabel('Income')
```

```python
# Verificando outliers utilizando age x loan
plt.scatter(base.iloc[:,2], base.iloc[:,3])
plt.xlabel('Age')
plt.ylabel('Loan')
```

```python
# Substituindo as idades < 0 pela média das idades válidas
base.loc[base.age < 0, 'age'] = base[base.age > 0].age.mean()
```

```python
# Verificando novamente os outliers income x age após manipulação das idades
plt.scatter(base.iloc[:,1], base.iloc[:,2])
plt.ylabel('Age')
plt.xlabel('Income')
```

```python
# Verificando novamente outliers utilizando age x loan após manipulação das idades
plt.scatter(base.iloc[:,2], base.iloc[:,3])
plt.xlabel('Age')
plt.ylabel('Loan')
```
